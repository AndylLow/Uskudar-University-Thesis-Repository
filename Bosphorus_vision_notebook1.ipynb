## Setup

### Subtask:
Mount Google Drive, locate the dataset, and install necessary libraries.

```python
!pip install ultralytics
```

```python
from google.colab import drive
import os
import torch
from ultralytics import YOLO

# Mount Google Drive
drive.mount('/content/drive')

# Define dataset path
dataset_path = "/content/drive/MyDrive/Uskudar Uni Master/YOLO Model klasor/Custom Dataset/bogaz_v_1.v1-roboflow-instant-1--eval-.yolov11"
data_yaml_path = os.path.join(dataset_path, 'data.yaml')

# Verify dataset location and content (optional, but good practice)
print(f"Listing contents of dataset path: {dataset_path}")
!ls "{dataset_path}"
if os.path.exists(data_yaml_path):
    print(f"\nContent of data.yaml:")
    with open(data_yaml_path, 'r') as f:
        print(f.read())
else:
    print(f"data.yaml not found at: {data_yaml_path}")


# Install ultralytics
%pip install ultralytics

# Check for GPU availability and set device
device = 'cuda' if torch.cuda.is_available() else 'cpu'
print(f"Using device: {device}")
```

## Load YOLOv11 Model

### Subtask:
Load a pre-trained YOLOv11 model.

**Reasoning**:
Load a pre-trained YOLOv11 model. As official 'yolov11.pt' might not be available, we will attempt to load a specific YOLOv11 model size, like 'yolov11m.pt' for medium. If this fails, we will fall back to 'yolov8n.pt' as an alternative.

```python
# Load a pre-trained YOLOv11 medium model (replace 'yolov11m.pt' if a different size is needed)
try:
    model = YOLO('yolo11m.pt')
    print("YOLOv11 medium model loaded successfully.")
except Exception as e:
    print(f"Failed to load yolo11m.pt: {e}")
```

## Configure and Fine-tune the Model

### Subtask:
Set up training parameters and start the fine-tuning process.

**Reasoning**:
Configure training parameters including dataset path, device, epochs, batch size, and image size. Start the training process using the loaded model and parameters. A smaller batch size is used due to the dataset size.

```python
import wandb
from ultralytics import YOLO

# Initialize Weights & Biases
wandb.init(project="YOLOv11-Fine-tuning", name="run-1")

# Define training parameters
epochs = 100
# Setting a smaller batch size as the dataset is small
batch_size = 8
imgsz = 640 # Image size

print(f"Dataset path: {data_yaml_path}")
print(f"Number of epochs: {epochs}")
print(f"Batch size: {batch_size}")
print(f"Image size: {imgsz}")
print(f"Using device: {device}")

# Start training
model.train(data=data_yaml_path, epochs=epochs, batch=batch_size, imgsz=imgsz, device=device)

# Finish W&B run
wandb.finish()
```

## Evaluate and Save the Model

### Subtask:
Evaluate the fine-tuned model and save the weights.

**Reasoning**:
Evaluate the performance of the trained model on the validation dataset and save the model weights.

```python
# Evaluate the model on the validation set
metrics = model.val(data=data_yaml_path, imgsz=imgsz, device=device)

# Print evaluation metrics
print("Evaluation Metrics:")
print(f"  mAP50: {metrics.results_dict['metrics/mAP50(B)']}")
print(f"  mAP50-95: {metrics.results_dict['metrics/mAP50-95(B)']}")

# Save the trained model weights
# The model weights are automatically saved in the 'runs' directory after training.
# You can find the weights in a directory structure similar to:
# /content/runs/detect/train/weights/best.pt
print("\nModel weights saved in the 'runs' directory.")
```

```python
# Define the path to the best weights of the trained model
best_weights_path = "/content/runs/detect/train/weights/best.pt"

# Define the path to the video file
video_path = "/content/drive/MyDrive/Uskudar Uni Master/YOLO Model klasor/CÌ§ekim Verileri/20250126_151515.mp4"

# Load the trained model
trained_model = YOLO(best_weights_path)

# Run inference on the video file
results = trained_model.predict(source=video_path, save=True, save_txt=True)

print("Inference complete. Results saved in the 'runs' directory.")
```
